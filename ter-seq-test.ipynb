```python
# рҹ“Ұ Imports
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
import os
import glob as gb
from sklearn.metrics import classification_report, confusion_matrix, balanced_accuracy_score, cohen_kappa_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (InputLayer, Conv2D, MaxPooling2D, Flatten, Dense, 
                                    Dropout, BatchNormalization, GlobalAveragePooling2D,
                                    Activation, SeparableConv2D, ReLU)
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from tensorflow.keras.optimizers import AdamW
from tensorflow.keras import mixed_precision
from tensorflow.keras.regularizers import l2

# вҡЎ Enable mixed precision
mixed_precision.set_global_policy('mixed_float16')

# вҡҷпёҸ Config
initial_lr = 1e-3  # Increased learning rate
IMG_SIZE = (299, 299)
BATCH_SIZE = 32
EPOCHS = 100  # Increased epochs
SEED = 42
DATA_DIR = "/kaggle/input/mpox-skin-lesion-dataset-version-20-msld-v20/Augmented Images/Augmented Images/FOLDS_AUG/fold3_AUG/Train"

# рҹ“Ӯ Load Datasets - First get class names
train_dataset_raw = tf.keras.utils.image_dataset_from_directory(
    DATA_DIR,
    shuffle=True,
    labels='inferred',
    batch_size=BATCH_SIZE,
    image_size=IMG_SIZE,
    validation_split=0.1,
    subset='training',
    seed=SEED
)

validation_dataset_raw = tf.keras.utils.image_dataset_from_directory(
    DATA_DIR,
    shuffle=True,
    labels='inferred',
    batch_size=BATCH_SIZE,
    image_size=IMG_SIZE,
    validation_split=0.1,
    subset='validation',
    seed=SEED
)

# Get class names before mapping
class_names = train_dataset_raw.class_names
num_classes = len(class_names)
print("Classes:", class_names)

# рҹ“Ҡ Visualize Class Distribution
data_imbalance = []
for class_folder in class_names:
    file_list = gb.glob(os.path.join(DATA_DIR, class_folder, '*.*'))
    data_imbalance.append(len(file_list))

plt.figure(figsize=(13, 6))
sns.barplot(x=class_names, y=data_imbalance, palette="rocket")
plt.title("Class Distribution")
plt.ylabel("Number of Images")
plt.xticks(rotation=15)
plt.show()

# рҹ“Ҳ Compute Class Weights
total_images = sum(data_imbalance)
class_weight = {i: (1.0 / count) * (total_images / len(data_imbalance)) for i, count in enumerate(data_imbalance)}
print("\nClass Weights:")
for i, w in class_weight.items():
    print(f"{class_names[i]}: {w:.2f}")

# рҹ”Ҙ Enhanced Data Augmentation
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal_and_vertical"),
    tf.keras.layers.RandomRotation(0.25),
    tf.keras.layers.RandomZoom(0.25),
    tf.keras.layers.RandomContrast(0.2),
    tf.keras.layers.RandomBrightness(0.2),
    tf.keras.layers.GaussianNoise(0.01),
    tf.keras.layers.RandomTranslation(0.15, 0.15)
])

# Preprocessing function
def preprocess(image, label):
    image = tf.image.convert_image_dtype(image, tf.float32)  # Normalize to [0,1]
    return image, label

# Apply preprocessing
train_dataset = train_dataset_raw.map(preprocess)
validation_dataset = validation_dataset_raw.map(preprocess)

# Optimize dataset pipeline
AUTOTUNE = tf.data.AUTOTUNE
train_dataset = train_dataset.cache().prefetch(buffer_size=AUTOTUNE)
validation_dataset = validation_dataset.cache().prefetch(buffer_size=AUTOTUNE)

# рҹҸ—пёҸ Build Optimized CNN Model with Residual Connections
def build_optimized_cnn(shape=(299, 299, 3), num_classes=6):
    model = Sequential([
        InputLayer(shape=shape),
        data_augmentation,
        
        # Initial Conv Block
        Conv2D(32, (7,7), strides=2, padding='same', kernel_regularizer=l2(1e-4)),
        BatchNormalization(),
        ReLU(),
        MaxPooling2D((3,3), strides=2, padding='same'),
        
        # Residual Block 1
        Conv2D(64, (3,3), padding='same', kernel_regularizer=l2(1e-4)),
        BatchNormalization(),
        ReLU(),
        Conv2D(64, (3,3), padding='same', kernel_regularizer=l2(1e-4)),
        BatchNormalization(),
        ReLU(),
        MaxPooling2D((2,2)),
        
        # Residual Block 2
        Conv2D(128, (3,3), padding='same', kernel_regularizer=l2(1e-4)),
        BatchNormalization(),
        ReLU(),
        Conv2D(128, (3,3), padding='same', kernel_regularizer=l2(1e-4)),
        BatchNormalization(),
        ReLU(),
        MaxPooling2D((2,2)),
        
        # Depthwise Separable Convolutions
        SeparableConv2D(256, (3,3), padding='same', 
                       depthwise_regularizer=l2(1e-4),
                       pointwise_regularizer=l2(1e-4)),
        BatchNormalization(),
        ReLU(),
        SeparableConv2D(256, (3,3), padding='same',
                       depthwise_regularizer=l2(1e-4),
                       pointwise_regularizer=l2(1e-4)),
        BatchNormalization(),
        ReLU(),
        MaxPooling2D((2,2)),
        
        # Final Feature Extraction
        Conv2D(512, (3,3), padding='same', kernel_regularizer=l2(1e-4)),
        BatchNormalization(),
        ReLU(),
        GlobalAveragePooling2D(),
        
        # Classifier Head
        Dense(512, activation='relu', kernel_regularizer=l2(1e-4)),
        Dropout(0.6),
        BatchNormalization(),
        Dense(256, activation='relu', kernel_regularizer=l2(1e-4)),
        Dropout(0.5),
        BatchNormalization(),
        Dense(num_classes, activation='softmax', dtype='float32')
    ])
    return model

model = build_optimized_cnn(shape=(299, 299, 3), num_classes=num_classes)

# вҡЎ Enhanced Compilation
optimizer = AdamW(learning_rate=initial_lr, weight_decay=1e-4)
model.compile(
    optimizer=optimizer,
    loss=tf.keras.losses.SparseCategoricalCrossentropy(),
    metrics=['accuracy', 
             tf.keras.metrics.SparseTopKCategoricalAccuracy(k=3, name='top3_accuracy')]
)

model.summary()

# вҸ° Enhanced Callbacks
class LRTracker(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs=None):
        logs = logs or {}
        logs['lr'] = tf.keras.backend.get_value(self.model.optimizer.learning_rate)

callbacks = [
    EarlyStopping(monitor='val_accuracy', patience=15, restore_best_weights=True, mode='max'),
    ReduceLROnPlateau(monitor='val_accuracy', factor=0.5, patience=5, verbose=1, mode='max'),
    ModelCheckpoint("optimized_cnn_best.keras", 
                   monitor="val_accuracy", 
                   save_best_only=True, 
                   mode='max',
                   verbose=1),
    LRTracker(),
    tf.keras.callbacks.TensorBoard(log_dir='./logs')
]

# рҹҸғвҖҚвҷӮпёҸ Train model
history = model.fit(
    train_dataset,
    validation_data=validation_dataset,
    class_weight=class_weight,
    epochs=EPOCHS,
    callbacks=callbacks,
    verbose=1
)

# рҹ§Ә Enhanced Evaluation
true_labels = []
predicted_probs = []

for images, labels in validation_dataset:
    batch_pred = model.predict(images, verbose=0)
    predicted_probs.extend(batch_pred)
    true_labels.extend(labels.numpy())

true_labels = np.array(true_labels)
predicted_probs = np.array(predicted_probs)
predicted_labels = np.argmax(predicted_probs, axis=1)

# рҹ“Ҳ Plot Metrics with more details
plt.figure(figsize=(18, 6))

# Accuracy plot
plt.subplot(1, 3, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
if 'top3_accuracy' in history.history:
    plt.plot(history.history['top3_accuracy'], label='Training Top-3 Accuracy')
if 'val_top3_accuracy' in history.history:
    plt.plot(history.history['val_top3_accuracy'], label='Validation Top-3 Accuracy')
plt.title('Accuracy Metrics')
plt.legend()

# Loss plot
plt.subplot(1, 3, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss Metrics')
plt.legend()

# Learning rate plot
plt.subplot(1, 3, 3)
if 'lr' in history.history:
    plt.plot(history.history['lr'], label='Learning Rate')
    plt.title('Learning Rate Schedule')
    plt.legend()
else:
    plt.text(0.5, 0.5, 'Learning Rate data not available', 
             ha='center', va='center')
    plt.title('Learning Rate Schedule')

plt.tight_layout()
plt.show()

# рҹ“Ӣ Generate and display Classification Report
print("\n" + "="*80)
print("Detailed Classification Report:")
print("="*80 + "\n")

clf_report = classification_report(
    true_labels,
    predicted_labels, 
    target_names=class_names,
    digits=4
)
print(clf_report)

# Additional performance metrics
print("\n" + "="*80)
print("Additional Performance Metrics:")
print("="*80 + "\n")

accuracy = np.mean(true_labels == predicted_labels)
balanced_acc = balanced_accuracy_score(true_labels, predicted_labels)
kappa = cohen_kappa_score(true_labels, predicted_labels)

print(f"Overall Accuracy: {accuracy:.4f}")
print(f"Balanced Accuracy: {balanced_acc:.4f}")
print(f"Cohen's Kappa Score: {kappa:.4f}")

# Plot normalized confusion matrix
plt.figure(figsize=(10, 8))
cm = confusion_matrix(true_labels, predicted_labels)
cm_norm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

sns.heatmap(cm_norm, 
            annot=True, 
            fmt=".2f", 
            cmap="Blues",
            xticklabels=class_names,
            yticklabels=class_names,
            cbar=False)

plt.title('Normalized Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
