```python
from ultralytics import YOLO

model = YOLO("yolo11n-cls.pt")  # or any other checkpoint

model.train(
    data="/kaggle/working/mpox_split_ready",  # one level higher
    task="classify",
    epochs=10,
    imgsz=640,
    batch=16,
    name="mpox_classify_fixed",
    project="/kaggle/working",  # ✅ Writable directory!
    plots=True
)
```
