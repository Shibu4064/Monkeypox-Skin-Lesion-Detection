```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers, callbacks, mixed_precision
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
from sklearn.preprocessing import label_binarize
from sklearn.model_selection import train_test_split
from itertools import cycle
import math
import os

# ⚙️ Configuration
CONFIG = {
    "img_size": (224, 224),
    "batch_size": 32,
    "epochs": 20,
    "seed": 42,
    "initial_lr": 1e-4,
    "warmup_epochs": 10,
    "weight_decay": 1e-4,
    "data_dir": "/kaggle/input/mpox-skin-lesion-dataset-version-20-msld-v20/Augmented Images/Augmented Images/FOLDS_AUG/fold3_AUG/Train",
    "validation_ratio": 0.2,
    "dropout_rate": 0.3,
    "embed_dim": 96,
    "num_heads": [3, 6, 12],
    "window_size": 7,
    "shift_size": 3,
    "num_mlp_units": [256, 128]
}

# Enable Mixed Precision
policy = mixed_precision.Policy('mixed_float16')
mixed_precision.set_global_policy(policy)

# 🖼️ Visualization Utilities
def show_sample_images(dataset, class_names, title="Sample Images", n=9):
    plt.figure(figsize=(15, 15))
    images, labels = next(iter(dataset.unbatch().batch(n)))
    for i in range(min(n, len(images))):
        ax = plt.subplot(math.ceil(n/3), 3, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(f"{class_names[labels[i]]}\nShape: {images[i].shape[:2]}", fontsize=10)
        plt.axis("off")
    plt.suptitle(title, y=1.02, fontsize=16)
    plt.tight_layout()
    plt.show()

# 📂 Data Loading with Enhanced Augmentation
def load_and_preprocess_data():
    full_ds = tf.keras.utils.image_dataset_from_directory(
        CONFIG["data_dir"],
        image_size=CONFIG["img_size"],
        batch_size=CONFIG["batch_size"],
        shuffle=False,
        seed=CONFIG["seed"]
    )

    class_names = full_ds.class_names
    num_classes = len(class_names)
    print(f"Found {num_classes} classes: {class_names}")

    # Get all labels and compute class weights
    all_labels = np.concatenate([y.numpy() for _, y in full_ds])
    unique_classes, counts = np.unique(all_labels, return_counts=True)
    
    # Stratified split
    train_idx, val_idx = train_test_split(
        np.arange(len(all_labels)),
        test_size=CONFIG["validation_ratio"],
        stratify=all_labels,
        random_state=CONFIG["seed"]
    )

    # Get all images
    all_images = np.concatenate([x.numpy() for x, _ in full_ds])

    # Create datasets
    def create_dataset(indices):
        return tf.data.Dataset.from_tensor_slices(
            (all_images[indices], all_labels[indices])
        ).batch(CONFIG["batch_size"]).prefetch(tf.data.AUTOTUNE)

    train_ds = create_dataset(train_idx)
    val_ds = create_dataset(val_idx)

    # Enhanced Augmentation Pipeline
    augmentation = tf.keras.Sequential([
        layers.RandomFlip("horizontal_and_vertical"),
        layers.RandomRotation(0.2),
        layers.RandomZoom(0.2),
        layers.RandomContrast(0.2),
        layers.RandomBrightness(0.2),
        layers.GaussianNoise(0.1),
    ])

    def preprocess(image, label):
        image = tf.image.convert_image_dtype(image, tf.float32)
        image = augmentation(image)
        image = tf.keras.applications.imagenet_utils.preprocess_input(image)
        return image, label

    train_ds = train_ds.map(preprocess, num_parallel_calls=tf.data.AUTOTUNE)
    val_ds = val_ds.map(
        lambda x, y: (tf.keras.applications.imagenet_utils.preprocess_input(tf.image.convert_image_dtype(x, tf.float32)), y),
        num_parallel_calls=tf.data.AUTOTUNE
    )

    return train_ds, val_ds, class_names, num_classes

# 🏗️ Swin Transformer Components (unchanged from original)
class WindowAttention(layers.Layer):
    def __init__(self, dim, num_heads, window_size, **kwargs):
        super().__init__(**kwargs)
        self.dim = dim
        self.num_heads = num_heads
        self.window_size = window_size
        self.scale = (dim // num_heads) ** -0.5
        
        self.qkv = layers.Dense(dim * 3)
        self.proj = layers.Dense(dim)
        
        # Relative position bias
        self.relative_position_bias_table = self.add_weight(
            shape=((2 * window_size - 1) * (2 * window_size - 1), num_heads),
            initializer="zeros",
            trainable=True
        )
        
        # Generate relative position indices
        coords = np.stack(np.meshgrid(
            range(window_size), range(window_size), indexing='ij'))
        coords = coords.reshape(2, -1)
        relative_coords = coords[:, :, None] - coords[:, None, :]
        relative_coords = relative_coords.transpose([1, 2, 0])
        relative_coords[:, :, 0] += window_size - 1
        relative_coords[:, :, 1] += window_size - 1
        relative_coords[:, :, 0] *= 2 * window_size - 1
        self.relative_position_indices = tf.constant(relative_coords.sum(-1).reshape(-1))
        
    def call(self, x):
        B, N, C = tf.shape(x)[0], tf.shape(x)[1], tf.shape(x)[2]
        
        qkv = self.qkv(x)
        qkv = tf.reshape(qkv, [B, N, 3, self.num_heads, C // self.num_heads])
        qkv = tf.transpose(qkv, [2, 0, 3, 1, 4])
        q, k, v = qkv[0], qkv[1], qkv[2]
        
        attn = (q @ tf.transpose(k, [0, 1, 3, 2])) * self.scale
        
        relative_position_bias = tf.gather(
            self.relative_position_bias_table,
            self.relative_position_indices
        )
        relative_position_bias = tf.reshape(
            relative_position_bias,
            [self.window_size * self.window_size, self.window_size * self.window_size, -1]
        )
        relative_position_bias = tf.transpose(relative_position_bias, [2, 0, 1])
        attn = attn + tf.expand_dims(relative_position_bias, 0)
        
        attn = tf.nn.softmax(attn, axis=-1)
        x = tf.transpose((attn @ v), [0, 2, 1, 3])
        x = tf.reshape(x, [B, N, C])
        x = self.proj(x)
        return x

class SwinTransformerBlock(layers.Layer):
    def __init__(self, dim, num_heads, window_size, shift_size=0, **kwargs):
        super().__init__(**kwargs)
        self.dim = dim
        self.num_heads = num_heads
        self.window_size = window_size
        self.shift_size = shift_size
        
        self.norm1 = layers.LayerNormalization(epsilon=1e-5)
        self.attn = WindowAttention(dim, num_heads, window_size)
        self.norm2 = layers.LayerNormalization(epsilon=1e-5)
        self.mlp = tf.keras.Sequential([
            layers.Dense(dim * 2, activation=tf.nn.gelu),
            layers.Dense(dim)
        ])
        
    def window_partition(self, x):
        B, H, W, C = tf.shape(x)[0], tf.shape(x)[1], tf.shape(x)[2], tf.shape(x)[3]
        x = tf.reshape(x, [B, H // self.window_size, self.window_size, W // self.window_size, self.window_size, C])
        x = tf.transpose(x, [0, 1, 3, 2, 4, 5])
        windows = tf.reshape(x, [-1, self.window_size, self.window_size, C])
        return windows
    
    def window_reverse(self, windows, H, W):
        B = tf.shape(windows)[0] // (H * W // self.window_size // self.window_size)
        x = tf.reshape(windows, [B, H // self.window_size, W // self.window_size, self.window_size, self.window_size, -1])
        x = tf.transpose(x, [0, 1, 3, 2, 4, 5])
        x = tf.reshape(x, [B, H, W, -1])
        return x
    
    def call(self, x):
        B, H, W, C = tf.shape(x)[0], tf.shape(x)[1], tf.shape(x)[2], tf.shape(x)[3]
        shortcut = x
        x = self.norm1(x)
        
        # Shifted window attention
        if self.shift_size > 0:
            shifted_x = tf.roll(x, shift=[-self.shift_size, -self.shift_size], axis=[1, 2])
        else:
            shifted_x = x
        
        # Window partition
        x_windows = self.window_partition(shifted_x)
        x_windows = tf.reshape(x_windows, [-1, self.window_size * self.window_size, C])
        
        # Attention
        attn_windows = self.attn(x_windows)
        attn_windows = tf.reshape(attn_windows, [-1, self.window_size, self.window_size, C])
        
        # Merge windows
        shifted_x = self.window_reverse(attn_windows, H, W)
        
        if self.shift_size > 0:
            x = tf.roll(shifted_x, shift=[self.shift_size, self.shift_size], axis=[1, 2])
        else:
            x = shifted_x
        
        # FFN
        x = shortcut + x
        x = x + self.mlp(self.norm2(x))
        return x

class PatchMerging(layers.Layer):
    def __init__(self, dim, **kwargs):
        super().__init__(**kwargs)
        self.dim = dim
        self.norm = layers.LayerNormalization(epsilon=1e-5)
        self.reduction = layers.Dense(2 * dim, use_bias=False)
    
    def call(self, x):
        B, H, W, C = tf.shape(x)[0], tf.shape(x)[1], tf.shape(x)[2], tf.shape(x)[3]
        
        x0 = x[:, 0::2, 0::2, :]
        x1 = x[:, 1::2, 0::2, :]
        x2 = x[:, 0::2, 1::2, :]
        x3 = x[:, 1::2, 1::2, :]
        x = tf.concat([x0, x1, x2, x3], axis=-1)
        x = tf.reshape(x, [B, H // 2, W // 2, 4 * C])
        x = self.norm(x)
        x = self.reduction(x)
        return x

# 🧠 Swin Transformer Model
def build_swin_model(num_classes):
    inputs = layers.Input(shape=(*CONFIG["img_size"], 3))
    
    # Initial patch embedding
    x = layers.Conv2D(
        CONFIG["embed_dim"],
        kernel_size=4,
        strides=4,
        padding='valid',
        kernel_regularizer=tf.keras.regularizers.l2(CONFIG["weight_decay"])
    )(inputs)
    x = layers.LayerNormalization(epsilon=1e-5)(x)
    
    # Swin stages
    for i, num_heads in enumerate(CONFIG["num_heads"]):
        # Swin blocks
        for _ in range(2):
            x = SwinTransformerBlock(
                dim=CONFIG["embed_dim"] * (2 ** i),
                num_heads=num_heads,
                window_size=CONFIG["window_size"],
                shift_size=0 if _ % 2 == 0 else CONFIG["window_size"] // 2
            )(x)
        
        # Patch merging (except last stage)
        if i < len(CONFIG["num_heads"]) - 1:
            x = PatchMerging(CONFIG["embed_dim"] * (2 ** i))(x)
    
    # Final layers
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.LayerNormalization(epsilon=1e-5)(x)
    
    # MLP head
    for units in CONFIG["num_mlp_units"]:
        x = layers.Dense(
            units,
            activation=tf.nn.gelu,
            kernel_regularizer=tf.keras.regularizers.l2(CONFIG["weight_decay"])
        )(x)
        x = layers.Dropout(CONFIG["dropout_rate"])(x)
    
    outputs = layers.Dense(
        num_classes,
        activation='softmax',
        dtype='float32',
        kernel_regularizer=tf.keras.regularizers.l2(CONFIG["weight_decay"])
    )(x)
    
    model = models.Model(inputs, outputs, name="Swin_Transformer")
    return model

# ⚙️ Training Setup with Warmup
def configure_training(model, num_classes):
    def lr_schedule(epoch):
        if epoch < CONFIG["warmup_epochs"]:
            return CONFIG["initial_lr"] * (epoch + 1) / CONFIG["warmup_epochs"]
        progress = (epoch - CONFIG["warmup_epochs"]) / (CONFIG["epochs"] - CONFIG["warmup_epochs"])
        return 0.5 * CONFIG["initial_lr"] * (1 + math.cos(math.pi * progress))
    
    optimizer = optimizers.AdamW(
        learning_rate=CONFIG["initial_lr"],
        weight_decay=CONFIG["weight_decay"]
    )

    model.compile(
        optimizer=optimizer,
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy',
                tf.keras.metrics.SparseCategoricalAccuracy(name='acc'),
                tf.keras.metrics.SparseTopKCategoricalAccuracy(k=3, name='top3_acc')]
    )

    return [
        callbacks.LearningRateScheduler(lr_schedule),
        callbacks.EarlyStopping(patience=10, restore_best_weights=True, monitor='val_acc'),
        callbacks.ModelCheckpoint('best_model.keras', save_best_only=True, monitor='val_acc'),
        callbacks.TensorBoard(log_dir='./logs'),
        callbacks.CSVLogger('training_log.csv'),
        callbacks.BackupAndRestore('./backup')
    ]

# 🚀 Main Execution
def main():
    # Clear session and setup GPU
    tf.keras.backend.clear_session()
    gpus = tf.config.list_physical_devices('GPU')
    if gpus:
        try:
            for gpu in gpus:
                tf.config.experimental.set_memory_growth(gpu, True)
        except RuntimeError as e:
            print(e)
    
    # Load data
    train_ds, val_ds, class_names, num_classes = load_and_preprocess_data()
    print(f"\nTraining with {num_classes} classes: {class_names}")
    show_sample_images(train_ds, class_names, "Augmented Training Samples")
    
    # Build model
    model = build_swin_model(num_classes)
    print("\nModel Summary:")
    model.summary()
    
    # Configure training
    callbacks_list = configure_training(model, num_classes)

    # Train
    print("\n=== Starting Training ===")
    history = model.fit(
        train_ds,
        validation_data=val_ds,
        epochs=CONFIG["epochs"],
        callbacks=callbacks_list,
        verbose=1
    )
    
    # Evaluation
    print("\n=== Evaluation ===")
    y_true = np.concatenate([y.numpy() for _, y in val_ds])
    y_pred = model.predict(val_ds, verbose=1)
    y_pred_classes = np.argmax(y_pred, axis=1)
    
    print("\nClassification Report:")
    print(classification_report(y_true, y_pred_classes, target_names=class_names, digits=4))
    
    # Confusion Matrix
    plt.figure(figsize=(10, 8))
    cm = confusion_matrix(y_true, y_pred_classes)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names)
    plt.title('Confusion Matrix')
    plt.savefig('confusion_matrix.png', bbox_inches='tight')
    plt.show()

    # ROC Curve (for binary/multi-class)
    if num_classes == 2:
        fpr, tpr, _ = roc_curve(y_true, y_pred[:, 1])
        roc_auc = auc(fpr, tpr)
        plt.figure(figsize=(8, 6))
        plt.plot(fpr, tpr, label=f'ROC curve (AUC = {roc_auc:.2f})')
    else:
        y_true_bin = label_binarize(y_true, classes=np.arange(num_classes))
        fpr, tpr, roc_auc = {}, {}, {}
        plt.figure(figsize=(10, 8))
        for i in range(num_classes):
            fpr[i], tpr[i], _ = roc_curve(y_true_bin[:, i], y_pred[:, i])
            roc_auc[i] = auc(fpr[i], tpr[i])
            plt.plot(fpr[i], tpr[i], label=f'{class_names[i]} (AUC = {roc_auc[i]:.2f})')
    
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend()
    plt.savefig('roc_curve.png', bbox_inches='tight')
    plt.show()

    # Training History
    plt.figure(figsize=(15, 5))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['acc'], label='Train')
    plt.plot(history.history['val_acc'], label='Validation')
    plt.title('Model Accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend()
    
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train')
    plt.plot(history.history['val_loss'], label='Validation')
    plt.title('Model Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend()
    
    plt.tight_layout()
    plt.savefig('training_history.png', bbox_inches='tight')
    plt.show()

    # Save final model
    model.save('swin_transformer_final.keras')
    print("\nModel saved successfully!")

if __name__ == "__main__":
    main()
```
