```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers, callbacks, mixed_precision
from tensorflow.keras.applications import Xception
from tensorflow.keras.applications.xception import preprocess_input
from tensorflow.keras.metrics import AUC
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
from sklearn.utils.class_weight import compute_class_weight
from sklearn.model_selection import train_test_split
import math
import os

# ⚙️ Configuration
CONFIG = {
    "img_size": (299, 299),  # Xception requires 299x299
    "batch_size": 32,
    "epochs_phase1": 15,     # Initial training (frozen base)
    "epochs_phase2": 15,     # Fine-tuning (unfrozen layers)
    "seed": 42,
    "initial_lr": 1e-4,
    "min_lr": 1e-7,
    "data_dir": "/kaggle/input/monkeypox-skin-lesion-dataset/Fold1/Fold1/Fold1",
    "validation_ratio": 0.2,
    "dropout_rate": 0.5,
    "weight_decay": 1e-4,
    "augmentation_intensity": 0.3
}

# Enable Mixed Precision
policy = mixed_precision.Policy('mixed_float16')
mixed_precision.set_global_policy(policy)

# 📂 Data Loading and Preprocessing (Modified to handle any number of classes)
def load_and_preprocess_data():
    full_ds = tf.keras.utils.image_dataset_from_directory(
        os.path.join(CONFIG["data_dir"], "Train"),  # Changed to use only Train folder
        image_size=CONFIG["img_size"],
        batch_size=CONFIG["batch_size"],
        shuffle=False,
        seed=CONFIG["seed"],
        color_mode='rgb'
    )

    class_names = full_ds.class_names
    num_classes = len(class_names)
    print(f"Found {num_classes} classes: {class_names}")

    # Get all labels and compute class weights
    all_labels = np.concatenate([y.numpy() for _, y in full_ds])
    class_weights = compute_class_weight('balanced', classes=np.unique(all_labels), y=all_labels)
    class_weights = dict(enumerate(class_weights))
    print(f"Class weights: {class_weights}")

    # Stratified split
    train_idx, val_idx = train_test_split(
        np.arange(len(all_labels)),
        test_size=CONFIG["validation_ratio"],
        stratify=all_labels,
        random_state=CONFIG["seed"]
    )

    # Get all images
    all_images = np.concatenate([x.numpy() for x, _ in full_ds])

    # Create datasets
    def create_dataset(indices):
        return tf.data.Dataset.from_tensor_slices(
            (all_images[indices], all_labels[indices])
        ).batch(CONFIG["batch_size"]).prefetch(tf.data.AUTOTUNE)

    train_ds = create_dataset(train_idx)
    val_ds = create_dataset(val_idx)

    # Enhanced Augmentation for Medical Images
    augmentation = tf.keras.Sequential([
        layers.RandomFlip("horizontal_and_vertical"),
        layers.RandomRotation(CONFIG["augmentation_intensity"]),
        layers.RandomZoom(CONFIG["augmentation_intensity"]),
        layers.RandomContrast(CONFIG["augmentation_intensity"]/2),
        layers.GaussianNoise(0.02),
        layers.RandomBrightness(0.1),
    ], name="augmentation")

    def preprocess(image, label):
        image = tf.image.convert_image_dtype(image, tf.float32)
        image = augmentation(image)
        image = preprocess_input(image)  # Xception specific preprocessing
        return image, label

    train_ds = train_ds.map(preprocess, num_parallel_calls=tf.data.AUTOTUNE)
    val_ds = val_ds.map(
        lambda x, y: (preprocess_input(tf.image.convert_image_dtype(x, tf.float32)), y),
        num_parallel_calls=tf.data.AUTOTUNE
    )

    return train_ds, val_ds, class_names, class_weights, num_classes

# 🧠 Modified Xception Model (Handles both binary and multi-class)
def build_xception_model(num_classes):
    base_model = Xception(
        include_top=False,
        input_shape=(*CONFIG["img_size"], 3),
        weights='imagenet'
    )
    base_model.trainable = False

    inputs = layers.Input(shape=(*CONFIG["img_size"], 3))
    x = base_model(inputs)
    x = layers.GlobalAveragePooling2D()(x)
    
    # Enhanced classifier head
    x = layers.Dense(1024, activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(CONFIG["dropout_rate"])(x)
    
    x = layers.Dense(512, activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(CONFIG["dropout_rate"]/2)(x)
    
    # Handle both binary and multi-class
    if num_classes == 2:
        outputs = layers.Dense(1, activation='sigmoid', dtype='float32')(x)
        loss = 'binary_crossentropy'
    else:
        outputs = layers.Dense(num_classes, activation='softmax', dtype='float32')(x)
        loss = 'categorical_crossentropy'
    
    model = models.Model(inputs, outputs)
    return model, base_model, loss

# ⚙️ Enhanced Callbacks
def get_callbacks():
    def cosine_decay(epoch):
        decay = 0.5 * (1 + math.cos(math.pi * epoch / (CONFIG["epochs_phase1"] + CONFIG["epochs_phase2"])))
        return max(CONFIG["initial_lr"] * decay, CONFIG["min_lr"])

    return [
        callbacks.LearningRateScheduler(cosine_decay, verbose=1),
        callbacks.EarlyStopping(
            patience=8,
            restore_best_weights=True,
            monitor='val_accuracy',
            mode='max',
            min_delta=0.001
        ),
        callbacks.ModelCheckpoint(
            'best_model.keras',
            save_best_only=True,
            monitor='val_accuracy',
            mode='max'
        ),
        callbacks.TensorBoard(log_dir='./logs'),
        callbacks.CSVLogger('training_log.csv'),
        callbacks.ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.5,
            patience=3,
            min_lr=CONFIG["min_lr"],
            verbose=1
        )
    ]

# 🚀 Main Execution
def main():
    # GPU setup
    tf.keras.backend.clear_session()
    gpus = tf.config.list_physical_devices('GPU')
    if gpus:
        try:
            for gpu in gpus:
                tf.config.experimental.set_memory_growth(gpu, True)
            logical_gpus = tf.config.list_logical_devices('GPU')
            print(f"{len(gpus)} Physical GPUs, {len(logical_gpus)} Logical GPUs")
        except RuntimeError as e:
            print(e)

    # Load data
    train_ds, val_ds, class_names, class_weights, num_classes = load_and_preprocess_data()
    print(f"\nClass names: {class_names}")
    print(f"Class weights: {class_weights}")

    # Build model
    model, base_model, loss = build_xception_model(num_classes)
    model.summary()

    # Phase 1: Train top layers
    print("\n=== Phase 1: Training Top Layers ===")
    model.compile(
        optimizer=optimizers.AdamW(
            learning_rate=CONFIG["initial_lr"],
            weight_decay=CONFIG["weight_decay"]
        ),
        loss=loss,
        metrics=['accuracy', AUC(name='auc', multi_label=num_classes > 2)]
    )

    history = model.fit(
        train_ds,
        validation_data=val_ds,
        epochs=CONFIG["epochs_phase1"],
        callbacks=get_callbacks(),
        class_weight=class_weights
    )

    # Phase 2: Fine-tune
    print("\n=== Phase 2: Fine-Tuning ===")
    base_model.trainable = True
    # Unfreeze top 50% of layers
    for layer in base_model.layers[:len(base_model.layers)//2]:
        layer.trainable = False

    model.compile(
        optimizer=optimizers.AdamW(
            learning_rate=CONFIG["initial_lr"]/10,
            weight_decay=CONFIG["weight_decay"]/10
        ),
        loss=loss,
        metrics=['accuracy', AUC(name='auc', multi_label=num_classes > 2)]
    )

    history_fine = model.fit(
        train_ds,
        validation_data=val_ds,
        initial_epoch=history.epoch[-1]+1,
        epochs=CONFIG["epochs_phase1"] + CONFIG["epochs_phase2"],
        callbacks=get_callbacks(),
        class_weight=class_weights
    )

    # Load best model
    model = tf.keras.models.load_model('best_model.keras')

    # Enhanced Evaluation
    print("\n=== Final Evaluation ===")
    y_true = np.concatenate([y.numpy() for _, y in val_ds])
    y_pred = model.predict(val_ds)
    
    if num_classes == 2:
        y_pred_classes = (y_pred > 0.5).astype(int)
    else:
        y_pred_classes = np.argmax(y_pred, axis=1)
    
    print("\nClassification Report:")
    print(classification_report(y_true, y_pred_classes, target_names=class_names, digits=4))
    
    # Confusion Matrix
    plt.figure(figsize=(8, 6))
    cm = confusion_matrix(y_true, y_pred_classes)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names)
    plt.title('Confusion Matrix', fontsize=14)
    plt.xlabel('Predicted', fontsize=12)
    plt.ylabel('Actual', fontsize=12)
    plt.show()

    # ROC Curve (only for binary)
    if num_classes == 2:
        fpr, tpr, _ = roc_curve(y_true, y_pred)
        roc_auc = auc(fpr, tpr)
        plt.figure(figsize=(8, 6))
        plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
        plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('Receiver Operating Characteristic')
        plt.legend(loc="lower right")
        plt.show()

    # Training History
    plt.figure(figsize=(15, 5))
    plt.subplot(1, 3, 1)
    plt.plot(history.history['accuracy'] + history_fine.history['accuracy'], label='Train')
    plt.plot(history.history['val_accuracy'] + history_fine.history['val_accuracy'], label='Validation')
    plt.title('Model Accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend()
    
    plt.subplot(1, 3, 2)
    plt.plot(history.history['loss'] + history_fine.history['loss'], label='Train')
    plt.plot(history.history['val_loss'] + history_fine.history['val_loss'], label='Validation')
    plt.title('Model Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend()
    
    plt.subplot(1, 3, 3)
    plt.plot(history.history['auc'] + history_fine.history['auc'], label='Train')
    plt.plot(history.history['val_auc'] + history_fine.history['val_auc'], label='Validation')
    plt.title('Model AUC')
    plt.ylabel('AUC')
    plt.xlabel('Epoch')
    plt.legend()
    
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    main()
```
