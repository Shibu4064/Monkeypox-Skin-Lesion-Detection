```python
# Improved SkinIncept Ensemble with Better Training Strategy
import tensorflow as tf
from tensorflow.keras.applications import InceptionV3, InceptionResNetV2, EfficientNetB0
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout, BatchNormalization
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.regularizers import l2
from sklearn.utils.class_weight import compute_class_weight
import numpy as np
import pandas as pd
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import os
import pickle

# Configuration
IMG_SIZE = 224
BATCH_SIZE = 16  # Reduced for better gradient updates
NUM_CLASSES = 6
EPOCHS = 50  # Reduced to prevent overfitting
LEARNING_RATE = 0.0001

# Set random seeds
tf.random.set_seed(42)
np.random.seed(42)

class ImprovedSkinInceptEnsemble:
    def __init__(self, img_size=224, num_classes=6):
        self.img_size = img_size
        self.num_classes = num_classes
        self.models = {}
        self.class_weights = None
        
    def calculate_class_weights(self, train_generator):
        """Calculate class weights to handle imbalanced dataset"""
        # Get class counts
        class_counts = np.bincount(train_generator.classes)
        total_samples = len(train_generator.classes)
        
        # Calculate weights
        class_weights = {}
        for i in range(len(class_counts)):
            class_weights[i] = total_samples / (len(class_counts) * class_counts[i])
        
        self.class_weights = class_weights
        print("Class weights calculated:")
        for class_idx, weight in class_weights.items():
            class_name = list(train_generator.class_indices.keys())[class_idx]
            print(f"{class_name}: {weight:.2f}")
        
        return class_weights
    
    def create_base_model(self, base_model_func, model_name, input_shape):
        """Create a base model with improved architecture"""
        # Load pre-trained model
        base_model = base_model_func(
            weights='imagenet',
            include_top=False,
            input_shape=input_shape
        )
        
        # Freeze base model initially
        base_model.trainable = False
        
        # Improved classification head
        x = base_model.output
        x = GlobalAveragePooling2D()(x)
        x = BatchNormalization()(x)
        
        # Add more regularization
        x = Dense(1024, activation='relu', kernel_regularizer=l2(0.01))(x)
        x = Dropout(0.5)(x)
        x = BatchNormalization()(x)
        
        x = Dense(512, activation='relu', kernel_regularizer=l2(0.01))(x)
        x = Dropout(0.4)(x)
        x = BatchNormalization()(x)
        
        x = Dense(256, activation='relu', kernel_regularizer=l2(0.01))(x)
        x = Dropout(0.3)(x)
        
        predictions = Dense(self.num_classes, activation='softmax', 
                          name=f'{model_name}_predictions')(x)
        
        model = Model(inputs=base_model.input, outputs=predictions)
        return model, base_model
    
    def build_models(self):
        """Build all models with different configurations"""
        input_shape = (self.img_size, self.img_size, 3)
        
        print("Building InceptionV3...")
        inception_v3, inception_v3_base = self.create_base_model(
            InceptionV3, 'InceptionV3', input_shape
        )
        self.models['InceptionV3'] = {
            'model': inception_v3,
            'base_model': inception_v3_base,
            'lr': LEARNING_RATE,
            'fine_tune_lr': LEARNING_RATE/20
        }
        
        print("Building InceptionResNetV2...")
        inception_resnet_v2, inception_resnet_v2_base = self.create_base_model(
            InceptionResNetV2, 'InceptionResNetV2', input_shape
        )
        self.models['InceptionResNetV2'] = {
            'model': inception_resnet_v2,
            'base_model': inception_resnet_v2_base,
            'lr': LEARNING_RATE * 0.5,  # Lower LR for this model
            'fine_tune_lr': LEARNING_RATE/30
        }
        
        print("Building EfficientNetB0...")
        efficientnet_b0, efficientnet_b0_base = self.create_base_model(
            EfficientNetB0, 'EfficientNetB0', input_shape
        )
        self.models['EfficientNetB0'] = {
            'model': efficientnet_b0,
            'base_model': efficientnet_b0_base,
            'lr': LEARNING_RATE * 0.7,  # Slightly lower LR
            'fine_tune_lr': LEARNING_RATE/25
        }
        
        print("All models built successfully!")
    
    def compile_models(self):
        """Compile models with different learning rates"""
        for model_name, model_dict in self.models.items():
            optimizer = Adam(learning_rate=model_dict['lr'])
            model_dict['model'].compile(
                optimizer=optimizer,
                loss='categorical_crossentropy',
                metrics=['accuracy']
            )
            print(f"{model_name} compiled with LR: {model_dict['lr']}")
    
    def get_callbacks(self, model_name):
        """Enhanced callbacks with better patience"""
        callbacks = [
            ModelCheckpoint(
                filepath=f'best_{model_name.lower()}_model.h5',
                monitor='val_accuracy',
                save_best_only=True,
                save_weights_only=False,
                verbose=1
            ),
            EarlyStopping(
                monitor='val_accuracy',
                patience=10,  # Reduced patience
                restore_best_weights=True,
                verbose=1
            ),
            ReduceLROnPlateau(
                monitor='val_loss',
                factor=0.3,  # More aggressive LR reduction
                patience=5,
                min_lr=1e-8,
                verbose=1
            )
        ]
        return callbacks
    
    def train_individual_models(self, train_generator, validation_generator):
        """Improved training with better strategies"""
        histories = {}
        
        for model_name, model_dict in self.models.items():
            print(f"\n{'='*60}")
            print(f"Training {model_name}")
            print(f"{'='*60}")
            
            model = model_dict['model']
            callbacks = self.get_callbacks(model_name)
            
            # Reset generators
            train_generator.reset()
            validation_generator.reset()
            
            # Phase 1: Train with frozen base model (longer)
            print(f"Phase 1: Training {model_name} with frozen base model...")
            try:
                history_phase1 = model.fit(
                    train_generator,
                    epochs=25,  # Increased epochs for better feature learning
                    validation_data=validation_generator,
                    callbacks=callbacks,
                    verbose=1,
                    steps_per_epoch=len(train_generator),
                    validation_steps=len(validation_generator),
                    class_weight=self.class_weights  # Add class weights
                )
            except Exception as e:
                print(f"Error in Phase 1 training for {model_name}: {e}")
                continue
            
            # Phase 2: Fine-tune with selective unfreezing
            print(f"Phase 2: Fine-tuning {model_name} with selective unfreezing...")
            
            # Unfreeze only the last few layers initially
            base_model = model_dict['base_model']
            for layer in base_model.layers[:-20]:  # Keep most layers frozen
                layer.trainable = False
            for layer in base_model.layers[-20:]:  # Unfreeze last 20 layers
                layer.trainable = True
            
            # Compile with lower learning rate
            model.compile(
                optimizer=Adam(learning_rate=model_dict['fine_tune_lr']),
                loss='categorical_crossentropy',
                metrics=['accuracy']
            )
            
            # Reset generators
            train_generator.reset()
            validation_generator.reset()
            
            try:
                history_phase2 = model.fit(
                    train_generator,
                    epochs=15,  # Fewer epochs for fine-tuning
                    validation_data=validation_generator,
                    callbacks=callbacks,
                    verbose=1,
                    steps_per_epoch=len(train_generator),
                    validation_steps=len(validation_generator),
                    class_weight=self.class_weights
                )
            except Exception as e:
                print(f"Error in Phase 2 training for {model_name}: {e}")
                histories[model_name] = history_phase1.history
                continue
            
            # Phase 3: Full fine-tuning with very low LR
            print(f"Phase 3: Full fine-tuning {model_name}...")
            
            # Unfreeze all layers
            base_model.trainable = True
            
            # Compile with even lower learning rate
            model.compile(
                optimizer=Adam(learning_rate=model_dict['fine_tune_lr']/5),
                loss='categorical_crossentropy',
                metrics=['accuracy']
            )
            
            # Reset generators
            train_generator.reset()
            validation_generator.reset()
            
            try:
                history_phase3 = model.fit(
                    train_generator,
                    epochs=10,  # Final fine-tuning
                    validation_data=validation_generator,
                    callbacks=callbacks,
                    verbose=1,
                    steps_per_epoch=len(train_generator),
                    validation_steps=len(validation_generator),
                    class_weight=self.class_weights
                )
            except Exception as e:
                print(f"Error in Phase 3 training for {model_name}: {e}")
                # Combine only phase 1 and 2
                combined_history = {}
                for key in history_phase1.history.keys():
                    combined_history[key] = history_phase1.history[key] + history_phase2.history[key]
                histories[model_name] = combined_history
                continue
            
            # Combine all histories
            combined_history = {}
            for key in history_phase1.history.keys():
                combined_history[key] = (history_phase1.history[key] + 
                                       history_phase2.history[key] + 
                                       history_phase3.history[key])
            
            histories[model_name] = combined_history
            
            # Load best weights
            try:
                model.load_weights(f'best_{model_name.lower()}_model.h5')
                print(f"Best weights loaded for {model_name}")
            except Exception as e:
                print(f"Warning: Could not load best weights for {model_name}: {e}")
            
            # Evaluate individual model
            train_generator.reset()
            validation_generator.reset()
            val_loss, val_acc = model.evaluate(validation_generator, verbose=0)
            print(f"{model_name} final validation accuracy: {val_acc:.4f}")
            
            print(f"{model_name} training completed!")
        
        return histories
    
    def create_smart_ensemble_predictions(self, test_generator):
        """Smart ensemble with model performance-based weighting"""
        predictions = {}
        model_accuracies = {}
        
        # Get predictions and calculate individual accuracies
        for model_name, model_dict in self.models.items():
            print(f"Getting predictions from {model_name}...")
            model = model_dict['model']
            
            test_generator.reset()
            pred = model.predict(test_generator, verbose=1)
            predictions[model_name] = pred
            
            # Calculate accuracy on test set
            test_generator.reset()
            y_true = test_generator.classes
            y_pred = np.argmax(pred, axis=1)
            accuracy = np.mean(y_pred == y_true)
            model_accuracies[model_name] = accuracy
            print(f"{model_name} individual accuracy: {accuracy:.4f}")
        
        # Calculate dynamic weights based on performance
        total_accuracy = sum(model_accuracies.values())
        if total_accuracy > 0:
            dynamic_weights = {name: acc/total_accuracy for name, acc in model_accuracies.items()}
        else:
            # Fallback to equal weights
            dynamic_weights = {name: 1/len(predictions) for name in predictions.keys()}
        
        print("\nDynamic ensemble weights:")
        for name, weight in dynamic_weights.items():
            print(f"{name}: {weight:.3f}")
        
        # Calculate weighted ensemble
        ensemble_pred = np.zeros_like(predictions[list(predictions.keys())[0]])
        for model_name, pred in predictions.items():
            ensemble_pred += dynamic_weights[model_name] * pred
        
        return ensemble_pred, predictions, model_accuracies
    
    def evaluate_ensemble(self, test_generator, class_names):
        """Enhanced evaluation with detailed analysis"""
        ensemble_pred, individual_predictions, model_accuracies = self.create_smart_ensemble_predictions(test_generator)
        
        # Get true labels
        test_generator.reset()
        y_true = test_generator.classes
        
        # Convert predictions to class labels
        y_pred_ensemble = np.argmax(ensemble_pred, axis=1)
        
        # Calculate ensemble accuracy
        ensemble_accuracy = np.mean(y_pred_ensemble == y_true)
        print(f"\nEnsemble Accuracy: {ensemble_accuracy:.4f}")
        print(f"Best Individual Model Accuracy: {max(model_accuracies.values()):.4f}")
        print(f"Improvement: {ensemble_accuracy - max(model_accuracies.values()):.4f}")
        
        # Detailed classification report
        print("\nDetailed Classification Report:")
        print(classification_report(y_true, y_pred_ensemble, target_names=class_names))
        
        # Class-wise analysis
        print("\nClass-wise Performance Analysis:")
        cm = confusion_matrix(y_true, y_pred_ensemble)
        for i, class_name in enumerate(class_names):
            class_accuracy = cm[i, i] / cm[i].sum() if cm[i].sum() > 0 else 0
            print(f"{class_name}: {class_accuracy:.3f} ({cm[i].sum()} samples)")
        
        # Enhanced confusion matrix
        plt.figure(figsize=(12, 8))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                   xticklabels=class_names, yticklabels=class_names,
                   cbar_kws={'label': 'Number of Samples'})
        plt.title('Improved Ensemble Model Confusion Matrix')
        plt.ylabel('True Label')
        plt.xlabel('Predicted Label')
        plt.tight_layout()
        plt.savefig('improved_ensemble_confusion_matrix.png', dpi=300, bbox_inches='tight')
        plt.show()
        
        return ensemble_accuracy, y_pred_ensemble, model_accuracies

# Enhanced data generators with better augmentation
def create_improved_data_generators(train_dir, test_dir, img_size=224, batch_size=16):
    """Create improved data generators"""
    
    # Enhanced training augmentation for better generalization
    train_datagen = ImageDataGenerator(
        rescale=1./255,
        rotation_range=30,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        brightness_range=[0.8, 1.2],
        validation_split=0.2
    )
    
    test_datagen = ImageDataGenerator(rescale=1./255)
    
    train_generator = train_datagen.flow_from_directory(
        train_dir,
        target_size=(img_size, img_size),
        batch_size=batch_size,
        class_mode='categorical',
        subset='training',
        shuffle=True,
        seed=42
    )
    
    validation_generator = train_datagen.flow_from_directory(
        train_dir,
        target_size=(img_size, img_size),
        batch_size=batch_size,
        class_mode='categorical',
        subset='validation',
        shuffle=False,
        seed=42
    )
    
    test_generator = test_datagen.flow_from_directory(
        test_dir,
        target_size=(img_size, img_size),
        batch_size=batch_size,
        class_mode='categorical',
        shuffle=False
    )
    
    return train_generator, validation_generator, test_generator

def main_improved():
    """Main function with improved pipeline"""
    
   
    # Set up paths (modify these according to your dataset structure)
    TRAIN_DIR = '/kaggle/input/mpox-skin-lesion-dataset-version-20-msld-v20/Original Images/Original Images/FOLDS/fold1/Train'  # Update with your train directory
    TEST_DIR = '/kaggle/input/mpox-skin-lesion-dataset-version-20-msld-v20/Original Images/Original Images/FOLDS/fold1/Test'    # Update with your test directory
    
    # Class names (modify according to your dataset)
    CLASS_NAMES = ['Chickenpox', 'Cowpox', 'HFMD', 'Healthy', 'Measles', 'Monkeypox'] 
    
    print("Setting up Improved SkinIncept Ensemble...")
    
    # Create improved data generators
    train_gen, val_gen, test_gen = create_improved_data_generators(
        TRAIN_DIR, TEST_DIR, IMG_SIZE, BATCH_SIZE
    )
    
    print(f"Training samples: {train_gen.samples}")
    print(f"Validation samples: {val_gen.samples}")
    print(f"Test samples: {test_gen.samples}")
    
    # Initialize improved ensemble
    ensemble = ImprovedSkinInceptEnsemble(img_size=IMG_SIZE, num_classes=NUM_CLASSES)
    
    # Calculate class weights
    ensemble.calculate_class_weights(train_gen)
    
    # Build and compile models
    ensemble.build_models()
    ensemble.compile_models()
    
    # Train models with improved strategy
    print("\nStarting improved training...")
    histories = ensemble.train_individual_models(train_gen, val_gen)
    
    # Evaluate ensemble
    print("\nEvaluating improved ensemble...")
    ensemble_accuracy, predictions, model_accuracies = ensemble.evaluate_ensemble(test_gen, CLASS_NAMES)
    
    return ensemble, histories, model_accuracies

if __name__ == "__main__":
    ensemble, histories, model_accuracies = main_improved()
```
