```python
# ✅ Final Working Version with Class Handling
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers, callbacks, mixed_precision
from tensorflow.keras.applications import ConvNeXtBase
from tensorflow.keras.applications.convnext import preprocess_input
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
from sklearn.preprocessing import label_binarize
from sklearn.utils.class_weight import compute_class_weight
from sklearn.model_selection import train_test_split
from itertools import cycle
import math

# ⚙️ Configuration
CONFIG = {
    "img_size": (224, 224),
    "batch_size": 16,
    "epochs": 15,
    "seed": 42,
    "initial_lr": 3e-5,
    "data_dir": "/kaggle/input/mpox-skin-lesion-dataset-version-20-msld-v20/Augmented Images/Augmented Images/FOLDS_AUG/fold3_AUG/Train",
    "validation_ratio": 0.2
}

# Enable Mixed Precision
policy = mixed_precision.Policy('mixed_float16')
mixed_precision.set_global_policy(policy)

# 🖼️ Visualization
def show_sample_images(dataset, class_names, title="Sample Images", n=9):
    plt.figure(figsize=(15, 15))
    images, labels = next(iter(dataset.unbatch().batch(n)))
    for i in range(min(n, len(images))):
        ax = plt.subplot(math.ceil(n/3), 3, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(f"{class_names[labels[i]]}\nShape: {images[i].shape[:2]}", fontsize=10)
        plt.axis("off")
    plt.suptitle(title, y=1.02, fontsize=16)
    plt.tight_layout()
    plt.show()

# 📂 Data Loading with Class Verification
def load_and_preprocess_data():
    full_ds = tf.keras.utils.image_dataset_from_directory(
        CONFIG["data_dir"],
        image_size=CONFIG["img_size"],
        batch_size=CONFIG["batch_size"],
        shuffle=False,
        seed=CONFIG["seed"]
    )

    class_names = full_ds.class_names
    num_classes = len(class_names)
    print(f"Found {num_classes} classes: {class_names}")

    # Get all labels and verify counts
    all_labels = np.concatenate([y.numpy() for _, y in full_ds])
    unique_classes = np.unique(all_labels)
    
    if len(unique_classes) != num_classes:
        raise ValueError(f"Dataset contains {len(unique_classes)} classes but expects {num_classes}")

    # Stratified split
    train_idx, val_idx = train_test_split(
        np.arange(len(all_labels)),
        test_size=CONFIG["validation_ratio"],
        stratify=all_labels,
        random_state=CONFIG["seed"]
    )

    # Get all images
    all_images = np.concatenate([x.numpy() for x, _ in full_ds])

    # Create datasets
    def create_dataset(indices):
        return tf.data.Dataset.from_tensor_slices(
            (all_images[indices], all_labels[indices])
        ).batch(CONFIG["batch_size"]).prefetch(tf.data.AUTOTUNE)

    train_ds = create_dataset(train_idx)
    val_ds = create_dataset(val_idx)

    # Data augmentation
    augmentation = tf.keras.Sequential([
        layers.RandomFlip("horizontal_and_vertical"),
        layers.RandomRotation(0.15),
        layers.RandomZoom(0.15),
        layers.RandomContrast(0.1),
    ])

    def preprocess(image, label):
        image = tf.image.convert_image_dtype(image, tf.float32)
        image = augmentation(image)
        return image, label

    train_ds = train_ds.map(preprocess, num_parallel_calls=tf.data.AUTOTUNE)
    val_ds = val_ds.map(
        lambda x, y: (tf.image.convert_image_dtype(x, tf.float32), y),
        num_parallel_calls=tf.data.AUTOTUNE
    )

    return train_ds, val_ds, class_names, num_classes

# 🧠 Model Building with Verified Output Shape
def build_convnext_model(num_classes):
    # Verify num_classes is valid
    if num_classes <= 1:
        raise ValueError(f"num_classes must be > 1, got {num_classes}")

    base_model = ConvNeXtBase(
        include_top=False,
        input_shape=(*CONFIG["img_size"], 3),
        weights='imagenet',
        pooling=None
    )
    
    # Freeze early layers
    for layer in base_model.layers[:100]:
        layer.trainable = False

    inputs = layers.Input(shape=(*CONFIG["img_size"], 3))
    x = preprocess_input(inputs)
    x = base_model(x)
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.LayerNormalization(epsilon=1e-6)(x)
    
    # Critical: Ensure output matches num_classes
    outputs = layers.Dense(num_classes, activation='softmax', dtype='float32')(x)

    model = models.Model(inputs, outputs, name=f"ConvNeXt_{num_classes}Classes")
    
    # Verify output shape
    if model.output_shape[-1] != num_classes:
        raise ValueError(f"Model output shape {model.output_shape} doesn't match num_classes {num_classes}")
    
    return model

# ⚙️ Training Setup
def configure_training(model, num_classes):
    optimizer = optimizers.AdamW(
        learning_rate=CONFIG["initial_lr"],
        weight_decay=1e-4
    )

    model.compile(
        optimizer=optimizer,
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )

    return [
        callbacks.EarlyStopping(patience=5, restore_best_weights=True, monitor='val_accuracy'),
        callbacks.ReduceLROnPlateau(factor=0.2, patience=3, monitor='val_loss'),
        callbacks.ModelCheckpoint('best_model.keras', save_best_only=True, monitor='val_accuracy'),
        callbacks.TensorBoard(log_dir='./logs'),
        callbacks.CSVLogger('training_log.csv')
    ]

# 🚀 Main Execution
def main():
    # Clear session and setup GPU
    tf.keras.backend.clear_session()
    gpus = tf.config.list_physical_devices('GPU')
    if gpus:
        try:
            for gpu in gpus:
                tf.config.experimental.set_memory_growth(gpu, True)
        except RuntimeError as e:
            print(e)
    
    # Load data
    train_ds, val_ds, class_names, num_classes = load_and_preprocess_data()
    print(f"\nTraining with {num_classes} classes: {class_names}")
    
    # Build model
    model = build_convnext_model(num_classes)
    print("\nModel Summary:")
    model.summary()
    
    # Verify output shape matches classes
    assert model.output_shape[-1] == num_classes, \
        f"Model output shape {model.output_shape} doesn't match num_classes {num_classes}"
    
    # Configure training
    callbacks_list = configure_training(model, num_classes)

    # Train
    print("\n=== Starting Training ===")
    try:
        history = model.fit(
            train_ds,
            validation_data=val_ds,
            epochs=CONFIG["epochs"],
            callbacks=callbacks_list,
            verbose=1
        )
    except Exception as e:
        print(f"\nTraining error: {e}")
        print("Possible solutions:")
        print("- Verify your dataset contains the expected number of classes")
        print("- Reduce batch size further (try 8 or 4)")
        print("- Check for class imbalance")
        return
    
    # Evaluation
    print("\n=== Evaluation ===")
    y_true = np.concatenate([y.numpy() for _, y in val_ds])
    y_pred = model.predict(val_ds, verbose=0)
    y_pred_classes = np.argmax(y_pred, axis=1)
    
    print("\nClassification Report:")
    print(classification_report(y_true, y_pred_classes, target_names=class_names, digits=4))
    
    print("\nConfusion Matrix:")
    plt.figure(figsize=(10, 8))
    sns.heatmap(confusion_matrix(y_true, y_pred_classes), 
                annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names)
    plt.title('Confusion Matrix')
    plt.show()

    # Plot training history
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train')
    plt.plot(history.history['val_accuracy'], label='Validation')
    plt.title('Model Accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend()
    
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train')
    plt.plot(history.history['val_loss'], label='Validation')
    plt.title('Model Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend()
    
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    main()
```
