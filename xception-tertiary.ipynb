```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers, callbacks, mixed_precision
from tensorflow.keras.applications import Xception
from tensorflow.keras.applications.xception import preprocess_input
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
from sklearn.preprocessing import label_binarize
from sklearn.utils.class_weight import compute_class_weight
from sklearn.model_selection import train_test_split
from itertools import cycle
import math
import os

# ⚙️ Configuration
CONFIG = {
    "img_size": (299, 299),  # Xception requires 299x299
    "batch_size": 32,        # Batch size
    "epochs": 30,            # Total epochs (phase 1 + phase 2)
    "seed": 42,
    "initial_lr": 1e-4,      # Initial learning rate
    "min_lr": 1e-7,          # Minimum learning rate
    "data_dir": "/kaggle/input/mpox-skin-lesion-dataset-version-20-msld-v20/Augmented Images/Augmented Images/FOLDS_AUG/fold3_AUG/Train",
    "validation_ratio": 0.2,
    "dropout_rate": 0.5,      # Dropout rate
    "weight_decay": 1e-4      # Weight decay
}

# Enable Mixed Precision
policy = mixed_precision.Policy('mixed_float16')
mixed_precision.set_global_policy(policy)

# 🖼️ Visualization
def show_sample_images(dataset, class_names, title="Sample Images", n=9):
    plt.figure(figsize=(15, 15))
    images, labels = next(iter(dataset.unbatch().batch(n)))
    for i in range(min(n, len(images))):
        ax = plt.subplot(math.ceil(n/3), 3, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(f"{class_names[labels[i]]}\nShape: {images[i].shape[:2]}", fontsize=10)
        plt.axis("off")
    plt.suptitle(title, y=1.02, fontsize=16)
    plt.tight_layout()
    plt.show()

# 📂 Data Loading and Preprocessing
def load_and_preprocess_data():
    full_ds = tf.keras.utils.image_dataset_from_directory(
        CONFIG["data_dir"],
        image_size=CONFIG["img_size"],
        batch_size=CONFIG["batch_size"],
        shuffle=False,
        seed=CONFIG["seed"]
    )

    class_names = full_ds.class_names
    num_classes = len(class_names)
    print(f"Found {num_classes} classes: {class_names}")

    # Get all labels
    all_labels = np.concatenate([y.numpy() for _, y in full_ds])
    unique_classes, counts = np.unique(all_labels, return_counts=True)
    
    # Compute class weights
    class_weights = compute_class_weight('balanced', classes=unique_classes, y=all_labels)
    class_weights = dict(zip(unique_classes, class_weights))
    print(f"Class weights: {class_weights}")
    
    # Stratified split
    train_idx, val_idx = train_test_split(
        np.arange(len(all_labels)),
        test_size=CONFIG["validation_ratio"],
        stratify=all_labels,
        random_state=CONFIG["seed"]
    )

    # Get all images
    all_images = np.concatenate([x.numpy() for x, _ in full_ds])

    # Create datasets
    def create_dataset(indices):
        return tf.data.Dataset.from_tensor_slices(
            (all_images[indices], all_labels[indices])
        ).batch(CONFIG["batch_size"]).prefetch(tf.data.AUTOTUNE)

    train_ds = create_dataset(train_idx)
    val_ds = create_dataset(val_idx)

    # Data Augmentation
    augmentation = tf.keras.Sequential([
        layers.RandomFlip("horizontal_and_vertical"),
        layers.RandomRotation(0.2),
        layers.RandomZoom(0.2),
        layers.RandomContrast(0.1),
        layers.GaussianNoise(0.01),
    ])

    def preprocess(image, label):
        image = tf.image.convert_image_dtype(image, tf.float32)
        image = augmentation(image)
        return image, label

    train_ds = train_ds.map(preprocess, num_parallel_calls=tf.data.AUTOTUNE)
    val_ds = val_ds.map(
        lambda x, y: (tf.image.convert_image_dtype(x, tf.float32), y),
        num_parallel_calls=tf.data.AUTOTUNE
    )

    return train_ds, val_ds, class_names, num_classes, class_weights

# 🧠 Xception Model
def build_xception_model(num_classes):
    base_model = Xception(
        include_top=False,
        input_shape=(*CONFIG["img_size"], 3),
        weights='imagenet'
    )
    base_model.trainable = False  # Freeze initially

    inputs = layers.Input(shape=(*CONFIG["img_size"], 3))
    x = preprocess_input(inputs)
    x = base_model(x)
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dropout(CONFIG["dropout_rate"])(x)
    outputs = layers.Dense(num_classes, activation='softmax', dtype='float32')(x)

    model = models.Model(inputs, outputs)
    return model, base_model

# ⚙️ Training Configuration
def get_callbacks():
    # Cosine LR decay callback
    def cosine_decay(epoch):
        decay = 0.5 * (1 + math.cos(math.pi * epoch / CONFIG["epochs"]))
        return CONFIG["initial_lr"] * decay

    return [
        callbacks.LearningRateScheduler(cosine_decay, verbose=1),
        callbacks.EarlyStopping(
            patience=10,
            restore_best_weights=True,
            monitor='val_accuracy',
            mode='max'
        ),
        callbacks.ModelCheckpoint(
            'best_model.keras',
            save_best_only=True,
            monitor='val_accuracy'
        ),
        callbacks.TensorBoard(log_dir='./logs'),
        callbacks.CSVLogger('training_log.csv')
    ]

# 🚀 Main Execution
def main():
    # GPU setup
    tf.keras.backend.clear_session()
    gpus = tf.config.list_physical_devices('GPU')
    if gpus:
        try:
            for gpu in gpus:
                tf.config.experimental.set_memory_growth(gpu, True)
        except RuntimeError as e:
            print(e)

    # Load data
    train_ds, val_ds, class_names, num_classes, class_weights = load_and_preprocess_data()
    print(f"\nTraining with {num_classes} classes: {class_names}")

    # Build model
    model, base_model = build_xception_model(num_classes)
    model.summary()

    # Phase 1: Train top layers
    print("\n=== Phase 1: Training Top Layers ===")
    model.compile(
        optimizer=optimizers.AdamW(learning_rate=CONFIG["initial_lr"]),
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )

    history = model.fit(
        train_ds,
        validation_data=val_ds,
        epochs=10,  # Initial training of top layers
        callbacks=get_callbacks()[:3],  # No LR decay in phase 1
        class_weight=class_weights
    )

    # Phase 2: Fine-tune
    print("\n=== Phase 2: Fine-Tuning ===")
    base_model.trainable = True
    for layer in base_model.layers[:100]:  # Freeze first 100 layers
        layer.trainable = False

    model.compile(
        optimizer=optimizers.AdamW(learning_rate=CONFIG["initial_lr"]/10),
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )

    history = model.fit(
        train_ds,
        validation_data=val_ds,
        initial_epoch=history.epoch[-1]+1,
        epochs=CONFIG["epochs"],
        callbacks=get_callbacks(),
        class_weight=class_weights
    )

    # Evaluation
    print("\n=== Evaluation ===")
    y_true = np.concatenate([y.numpy() for _, y in val_ds])
    y_pred = model.predict(val_ds)
    y_pred_classes = np.argmax(y_pred, axis=1)
    
    print("\nClassification Report:")
    print(classification_report(y_true, y_pred_classes, target_names=class_names, digits=4))
    
    # Confusion Matrix
    plt.figure(figsize=(10, 8))
    cm = confusion_matrix(y_true, y_pred_classes)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names)
    plt.title('Confusion Matrix')
    plt.show()

    # Training History
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train')
    plt.plot(history.history['val_accuracy'], label='Validation')
    plt.title('Model Accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend()
    
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train')
    plt.plot(history.history['val_loss'], label='Validation')
    plt.title('Model Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend()
    
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    main()
```
