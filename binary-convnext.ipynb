```python
# вң… Final Corrected ConvNeXt for Monkeypox (Fixed ROC Curve Calculation)

import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers, callbacks, mixed_precision
from tensorflow.keras.applications import ConvNeXtBase
from tensorflow.keras.applications.convnext import preprocess_input
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
from sklearn.preprocessing import label_binarize
from sklearn.model_selection import train_test_split
from sklearn.utils.class_weight import compute_class_weight
import os, math
from itertools import cycle

# вҡҷпёҸ Configuration
CONFIG = {
    "img_size": (224, 224),
    "batch_size": 16,
    "epochs": 10,
    "seed": 42,
    "initial_lr": 3e-5,
    "fine_tune_lr": 1e-6,
    "data_dir": "/kaggle/input/monkeypox-skin-lesion-dataset/Augmented Images/Augmented Images",
    "validation_ratio": 0.2,
    "patience": 8,
    "dropout_rate": 0.3,
    "weight_decay": 1e-4,
    "label_smoothing": 0.1,
    "fine_tune_at": 100
}

# Mixed Precision
policy = mixed_precision.Policy('mixed_float16')
mixed_precision.set_global_policy(policy)

# рҹ“Ӯ Data Pipeline
def load_and_preprocess_data():
    full_ds = tf.keras.utils.image_dataset_from_directory(
        CONFIG["data_dir"],
        image_size=CONFIG["img_size"],
        batch_size=CONFIG["batch_size"],
        shuffle=False,
        seed=CONFIG["seed"]
    )
    class_names = full_ds.class_names
    num_classes = len(class_names)

    # Get all data for proper stratification
    all_labels = np.concatenate([y.numpy() for _, y in full_ds])
    all_images = np.concatenate([x.numpy() for x, _ in full_ds])
    
    # Stratified split
    train_idx, val_idx = train_test_split(
        np.arange(len(all_labels)),
        test_size=CONFIG["validation_ratio"],
        stratify=all_labels,
        random_state=CONFIG["seed"]
    )

    # Graph-compatible augmentation
    def augment_image(image, label):
        image = tf.image.convert_image_dtype(image, tf.float32)
        image = tf.image.random_flip_left_right(image)
        image = tf.image.random_flip_up_down(image)
        image = tf.image.random_brightness(image, 0.1)
        image = tf.image.random_contrast(image, 0.9, 1.1)
        
        # Rotation using tf.image.rot90
        rotations = tf.random.uniform(shape=[], minval=0, maxval=4, dtype=tf.int32)
        image = tf.image.rot90(image, k=rotations)
        
        # Zoom using central crop
        zoom_factor = tf.random.uniform([], 0.85, 1.0)
        new_size = tf.cast(tf.cast(tf.shape(image)[:2], tf.float32) * zoom_factor, tf.int32)
        image = tf.image.resize(image, new_size)
        image = tf.image.resize_with_crop_or_pad(image, *CONFIG["img_size"])
        
        return preprocess_input(image), tf.one_hot(label, depth=num_classes)

    def preprocess_val(image, label):
        return preprocess_input(tf.image.convert_image_dtype(image, tf.float32)), tf.one_hot(label, depth=num_classes)

    # Create datasets
    train_ds = (tf.data.Dataset.from_tensor_slices((all_images[train_idx], all_labels[train_idx]))
                .map(augment_image, num_parallel_calls=tf.data.AUTOTUNE)
                .batch(CONFIG["batch_size"])
                .prefetch(tf.data.AUTOTUNE))
    
    val_ds = (tf.data.Dataset.from_tensor_slices((all_images[val_idx], all_labels[val_idx]))
              .map(preprocess_val, num_parallel_calls=tf.data.AUTOTUNE)
              .batch(CONFIG["batch_size"])
              .prefetch(tf.data.AUTOTUNE))

    return train_ds, val_ds, class_names, num_classes, all_labels[train_idx]

# рҹ§  Model
def build_model(num_classes):
    base_model = ConvNeXtBase(
        include_top=False,
        input_shape=(*CONFIG["img_size"], 3),
        weights='imagenet',
        pooling='avg'
    )
    
    # Freeze initial layers
    for layer in base_model.layers[:CONFIG["fine_tune_at"]]:
        layer.trainable = False
    
    inputs = layers.Input(shape=(*CONFIG["img_size"], 3))
    x = base_model(inputs)
    
    # Enhanced classification head
    x = layers.LayerNormalization(epsilon=1e-6)(x)
    x = layers.Dropout(CONFIG["dropout_rate"])(x)
    x = layers.Dense(512, activation='gelu')(x)
    x = layers.LayerNormalization(epsilon=1e-6)(x)
    x = layers.Dropout(CONFIG["dropout_rate"]*0.5)(x)
    
    outputs = layers.Dense(
        num_classes, 
        activation='softmax',
        dtype='float32'
    )(x)
    
    return models.Model(inputs, outputs)

# рҹ§° Training Configuration
def configure_training(model, y_train, num_classes):
    class_weights = compute_class_weight(
        'balanced', 
        classes=np.unique(y_train), 
        y=y_train
    )
    class_weight_dict = {i: w for i, w in enumerate(class_weights)}
    
    # Create a new optimizer with fixed learning rate for fine-tuning compatibility
    optimizer = optimizers.AdamW(
        learning_rate=CONFIG["initial_lr"],
        weight_decay=CONFIG["weight_decay"],
        global_clipnorm=1.0
    )
    
    # Using CategoricalCrossentropy with one-hot labels
    loss = tf.keras.losses.CategoricalCrossentropy(
        label_smoothing=CONFIG["label_smoothing"]
    )
    
    # Callbacks
    callbacks_list = [
        callbacks.EarlyStopping(
            monitor='val_accuracy',
            patience=CONFIG["patience"],
            restore_best_weights=True,
            min_delta=0.001
        ),
        callbacks.ModelCheckpoint(
            'best_model.keras',
            monitor='val_accuracy',
            save_best_only=True,
            save_weights_only=False
        ),
        callbacks.ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.5,
            patience=3,
            min_lr=1e-7
        ),
        callbacks.TensorBoard(log_dir='./logs'),
        callbacks.CSVLogger('training_log.csv')
    ]
    
    model.compile(
        optimizer=optimizer,
        loss=loss,
        metrics=['accuracy',
                tf.keras.metrics.Precision(name='precision'),
                tf.keras.metrics.Recall(name='recall'),
                tf.keras.metrics.AUC(name='auc', multi_label=True)]  # Enable multi-class AUC
    )
    
    return callbacks_list, class_weight_dict

# рҹ”„ Fine-tuning
def fine_tune_model(model, train_ds, val_ds, callbacks_list, class_weight):
    # Unfreeze more layers
    for layer in model.layers[1].layers[CONFIG["fine_tune_at"]:]:
        layer.trainable = True
    
    # Create a new optimizer with lower learning rate for fine-tuning
    optimizer = optimizers.AdamW(
        learning_rate=CONFIG["fine_tune_lr"],
        weight_decay=CONFIG["weight_decay"]*0.1,
        global_clipnorm=1.0
    )
    
    model.compile(
        optimizer=optimizer,
        loss=tf.keras.losses.CategoricalCrossentropy(
            label_smoothing=CONFIG["label_smoothing"]*0.5
        ),
        metrics=['accuracy']
    )
    
    print("\nStarting fine-tuning...")
    history = model.fit(
        train_ds,
        validation_data=val_ds,
        epochs=CONFIG["epochs"],
        initial_epoch=CONFIG["epochs"]//2,
        callbacks=callbacks_list,
        class_weight=class_weight
    )
    return history

# рҹ“Ҡ Evaluation (Fixed ROC Curve)
def evaluate_model(model, val_ds, class_names):
    # Convert validation dataset back to original labels for evaluation
    val_ds_original = val_ds.map(lambda x, y: (x, tf.argmax(y, axis=1)))
    
    y_true = []
    y_pred = []
    y_score = []
    
    for images, labels in val_ds_original:
        y_true.extend(labels.numpy())
        preds = model.predict(images, verbose=0)
        y_pred.extend(np.argmax(preds, axis=1))
        y_score.extend(preds)
    
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    y_score = np.array(y_score)
    
    print("\n=== Classification Report ===")
    print(classification_report(y_true, y_pred, target_names=class_names, digits=4))
    
    print("\n=== Confusion Matrix ===")
    plt.figure(figsize=(10, 8))
    sns.heatmap(confusion_matrix(y_true, y_pred), 
                annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names)
    plt.title('Confusion Matrix')
    plt.show()
    
    # ROC Curve calculation
    print("\n=== ROC Curve ===")
    n_classes = len(class_names)
    
    if n_classes == 2:
        # Binary classification
        fpr, tpr, _ = roc_curve(y_true, y_score[:, 1])
        roc_auc = auc(fpr, tpr)
        
        plt.figure(figsize=(8, 6))
        plt.plot(fpr, tpr, color='darkorange', lw=2,
                label=f'ROC curve (AUC = {roc_auc:.2f})')
        plt.plot([0, 1], [0, 1], 'k--', lw=2)
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('ROC Curve')
        plt.legend(loc="lower right")
        plt.show()
    else:
        # Multi-class classification
        y_true_bin = label_binarize(y_true, classes=np.arange(n_classes))
        fpr, tpr, roc_auc = {}, {}, {}
        
        for i in range(n_classes):
            if len(np.unique(y_true_bin[:, i])) > 1:  # Check if class has both positive and negative samples
                fpr[i], tpr[i], _ = roc_curve(y_true_bin[:, i], y_score[:, i])
                roc_auc[i] = auc(fpr[i], tpr[i])
        
        plt.figure(figsize=(10, 8))
        colors = cycle(['aqua', 'darkorange', 'cornflowerblue', 'green', 'red', 'purple'])
        
        for i, color in zip(range(n_classes), colors):
            if i in roc_auc:  # Only plot if ROC was calculated for this class
                plt.plot(fpr[i], tpr[i], color=color,
                        label=f'{class_names[i]} (AUC = {roc_auc[i]:.2f})')
        
        plt.plot([0, 1], [0, 1], 'k--')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('Multi-class ROC Curve')
        plt.legend(loc="lower right")
        plt.show()
# рҹҡҖ Main Function
def main():
    # GPU setup
    gpus = tf.config.list_physical_devices('GPU')
    if gpus:
        try:
            for gpu in gpus:
                tf.config.experimental.set_memory_growth(gpu, True)
        except RuntimeError as e:
            print(e)
    
    tf.keras.backend.clear_session()
    
    # Load data
    train_ds, val_ds, class_names, num_classes, y_train = load_and_preprocess_data()
    
    # Build model
    model = build_model(num_classes)
    model.summary()
    
    # Configure training
    callbacks_list, class_weight = configure_training(model, y_train, num_classes)
    
    # Initial training
    print("\n=== Initial Training ===")
    history = model.fit(
        train_ds,
        validation_data=val_ds,
        epochs=CONFIG["epochs"]//2,
        callbacks=callbacks_list,
        class_weight=class_weight
    )
    
    # Fine-tune if validation accuracy < 95%
    best_acc = max(history.history['val_accuracy'])
    if best_acc < 0.95:
        history = fine_tune_model(model, train_ds, val_ds, callbacks_list, class_weight)
    
    # Final evaluation
    evaluate_model(model, val_ds, class_names)
    
    # Save model
    model.save('final_convnext_model.keras')
    print("\nModel saved as 'final_convnext_model.keras'")

if __name__ == "__main__":
    main()
```
