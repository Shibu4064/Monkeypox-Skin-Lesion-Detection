```python
# рҹ“Ұ Optimized Imports
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers, callbacks, initializers
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.utils.class_weight import compute_class_weight
from sklearn.model_selection import train_test_split

# вҡҷпёҸ Configuration
CONFIG = {
    "img_size": (224, 224),
    "batch_size": 32,
    "epochs": 50,
    "seed": 42,
    "initial_lr": 1e-4,
    "data_dir": "/kaggle/input/mpox-skin-lesion-dataset-version-20-msld-v20/Augmented Images/Augmented Images/FOLDS_AUG/fold3_AUG/Train",
    "validation_ratio": 0.2  # New parameter for stratified split
}

# рҹ“Ӯ Enhanced Data Pipeline with Stratified Sampling
def load_and_preprocess_data():
    # Load full dataset without splitting first
    full_ds = tf.keras.utils.image_dataset_from_directory(
        CONFIG["data_dir"],
        image_size=CONFIG["img_size"],
        batch_size=CONFIG["batch_size"],
        shuffle=False,  # Important for consistent stratification
        seed=CONFIG["seed"]
    )
    
    # Get class names and raw labels
    class_names = full_ds.class_names
    print(f"Classes: {class_names}")
    
    # Extract all labels and image paths for stratification
    all_images = []
    all_labels = []
    for images, labels in full_ds:
        all_images.extend(images.numpy())
        all_labels.extend(labels.numpy())
    
    # Convert to numpy arrays
    all_images = np.array(all_images)
    all_labels = np.array(all_labels)
    
    # Stratified train-test split
    train_idx, val_idx = train_test_split(
        np.arange(len(all_labels)),
        test_size=CONFIG["validation_ratio"],
        stratify=all_labels,
        random_state=CONFIG["seed"]
    )
    
    # Create TensorFlow datasets from the indices
    def create_dataset(indices):
        ds = tf.data.Dataset.from_tensor_slices(
            (all_images[indices], all_labels[indices]))
        return ds.batch(CONFIG["batch_size"])
    
    train_ds = create_dataset(train_idx)
    val_ds = create_dataset(val_idx)
    
    # Check class distribution
    def get_class_distribution(dataset):
        class_counts = np.zeros(len(class_names))
        for _, labels in dataset:
            labels = labels.numpy()
            for class_idx in range(len(class_names)):
                class_counts[class_idx] += np.sum(labels == class_idx)
        return class_counts
    
    train_counts = get_class_distribution(train_ds)
    val_counts = get_class_distribution(val_ds)
    
    print("\nTraining set class distribution:")
    for name, count in zip(class_names, train_counts):
        print(f"{name}: {count} samples ({count/sum(train_counts):.1%})")
    
    print("\nValidation set class distribution:")
    for name, count in zip(class_names, val_counts):
        print(f"{name}: {count} samples ({count/sum(val_counts):.1%})")
    
    # Data augmentation
    augmentation = tf.keras.Sequential([
        layers.RandomFlip("horizontal_and_vertical"),
        layers.RandomRotation(0.15),
        layers.RandomZoom(0.15),
        layers.RandomContrast(0.1),
    ])
    
    def preprocess(image, label):
        image = tf.image.convert_image_dtype(image, tf.float32)
        image = augmentation(image)
        return image, label
    
    # Apply preprocessing
    train_ds = train_ds.map(preprocess, num_parallel_calls=tf.data.AUTOTUNE)
    val_ds = val_ds.map(lambda x, y: (tf.image.convert_image_dtype(x, tf.float32), y),
                      num_parallel_calls=tf.data.AUTOTUNE)
    
    return train_ds.cache().prefetch(tf.data.AUTOTUNE), val_ds.cache().prefetch(tf.data.AUTOTUNE), class_names, train_counts

# рҹҸ—пёҸ Model Architecture (Same as before)
def build_custom_hornet(num_classes, class_counts=None):
    inputs = layers.Input(shape=(*CONFIG["img_size"], 3))
    
    # Stem
    x = layers.Conv2D(64, (7, 7), strides=2, padding='same')(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('gelu')(x)
    x = layers.MaxPooling2D((3, 3), strides=2, padding='same')(x)
    
    # Hierarchical stages
    for filters in [128, 256, 512]:
        # Residual block
        shortcut = x
        x = layers.GroupNormalization(groups=32)(x)
        x = layers.Conv2D(filters, (3, 3), padding='same')(x)
        x = layers.BatchNormalization()(x)
        x = layers.Activation('gelu')(x)
        
        if shortcut.shape[-1] != filters:
            shortcut = layers.Conv2D(filters, (1, 1))(shortcut)
        
        x = layers.Add()([shortcut, x])
        x = layers.MaxPooling2D((2, 2))(x)
    
    # Head
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dense(1024, activation='gelu')(x)
    x = layers.Dropout(0.5)(x)
    
    # Initialize output layer with class balancing
    if class_counts is not None:
        bias_init = initializers.Constant(np.log(class_counts + 1e-7))
    else:
        bias_init = 'zeros'
        
    outputs = layers.Dense(
        num_classes, 
        activation='softmax',
        bias_initializer=bias_init
    )(x)
    
    return models.Model(inputs, outputs)

# вҡЎ Training Setup (Same as before)
def configure_training(model, train_counts):
    # Compute class weights
    class_weights = compute_class_weight(
        'balanced',
        classes=np.arange(len(train_counts)),
        y=np.repeat(np.arange(len(train_counts)), train_counts.astype(int))
    )
    class_weights = dict(enumerate(class_weights))
    
    optimizer = optimizers.AdamW(
        learning_rate=CONFIG["initial_lr"],
        weight_decay=1e-4
    )
    
    model.compile(
        optimizer=optimizer,
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )
    
    return [
        callbacks.EarlyStopping(patience=10, restore_best_weights=True),
        callbacks.ReduceLROnPlateau(factor=0.5, patience=3),
        callbacks.ModelCheckpoint('best_model.keras', save_best_only=True),
        callbacks.TensorBoard(log_dir='./logs')
    ], class_weights

# рҹ“Ҡ Enhanced Evaluation Function
def evaluate_model(model, val_ds, class_names):
    y_true, y_pred = [], []
    for images, labels in val_ds:
        y_true.extend(labels.numpy())
        y_pred.extend(np.argmax(model.predict(images, verbose=0), axis=1))
    
    # Only show classes that exist in y_true
    present_classes = np.unique(y_true)
    present_class_names = [class_names[i] for i in present_classes]
    
    print("\nClassification Report:")
    print(classification_report(
        y_true,
        y_pred,
        labels=present_classes,
        target_names=present_class_names,
        digits=4,
        zero_division=0
    ))
    
    # Confusion matrix
    plt.figure(figsize=(10, 8))
    cm = confusion_matrix(y_true, y_pred, labels=present_classes)
    sns.heatmap(
        cm,
        annot=True,
        fmt='d',
        cmap='Blues',
        xticklabels=present_class_names,
        yticklabels=present_class_names
    )
    plt.title('Confusion Matrix')
    plt.show()

    # Detailed class distribution
    print("\nValidation Set Class Distribution:")
    for i, name in enumerate(class_names):
        count = np.sum(np.array(y_true) == i)
        if count > 0:  # Only show present classes
            print(f"{name}: {count} samples ({count/len(y_true):.1%})")

# рҹ“Ҳ Training Visualization (Same as before)
def plot_training_curves(history):
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train')
    plt.plot(history.history['val_accuracy'], label='Validation')
    plt.title('Accuracy')
    plt.legend()
    
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train')
    plt.plot(history.history['val_loss'], label='Validation')
    plt.title('Loss')
    plt.legend()
    
    plt.tight_layout()
    plt.show()

# рҹҡҖ Main Execution
def main():
    # Load data with stratified sampling
    train_ds, val_ds, class_names, train_counts = load_and_preprocess_data()
    
    # Build model
    model = build_custom_hornet(len(class_names), train_counts)
    model.summary()
    
    # Configure training
    callbacks_list, class_weights = configure_training(model, train_counts)
    
    # Train with class weights
    history = model.fit(
        train_ds,
        validation_data=val_ds,
        epochs=CONFIG["epochs"],
        callbacks=callbacks_list,
        class_weight=class_weights,
        verbose=1
    )
    
    # Evaluate
    evaluate_model(model, val_ds, class_names)
    
    # Plot training curves
    plot_training_curves(history)

if __name__ == "__main__":
    main()
```
